# Práctica — Ejercicio 9: Representación de datos en múltiples formatos

Representar la siguiente tabla de productos en los formatos: **CSV** (delimitador `|`),
**TXT**, **YAML**, **XML**, **JSON** y **HTML**.

| id | desc_prod | precio | proveedor |
|---|---|---|---|
| 0049570 | camisa | 2000 | fashionistas |
| 0769298 | jean | 6000 | tu moda |
| 8458909 | polera | 3000 | el ropero |

---

```python
import pandas as pd
import json
import yaml
import xml.etree.ElementTree as ET
from pathlib import Path

# Tabla de productos
productos = [
    {'id': '0049570', 'desc_prod': 'camisa',  'precio': 2000, 'proveedor': 'fashionistas'},
    {'id': '0769298', 'desc_prod': 'jean',    'precio': 6000, 'proveedor': 'tu moda'},
    {'id': '8458909', 'desc_prod': 'polera',  'precio': 3000, 'proveedor': 'el ropero'},
]
df = pd.DataFrame(productos)
print(df.to_string(index=False))

OUTDIR = Path('.')
```

## Ítem 1 — Escritura en cada formato

> **Represente la tabla en CSV (delimitador `|`), TXT, YAML, XML, JSON y HTML.**

```python
# ── CSV con | como delimitador ──────────────────────────────
csv_path = OUTDIR / 'productos.csv'
df.to_csv(csv_path, sep='|', index=False)
print("=== CSV ===")
print(csv_path.read_text())
```

```python
# ── TXT (alineado con espacios) ─────────────────────────────
txt_path = OUTDIR / 'productos.txt'
txt_path.write_text(df.to_string(index=False))
print("=== TXT ===")
print(txt_path.read_text())
```

```python
# ── YAML ────────────────────────────────────────────────────
yaml_path = OUTDIR / 'productos.yaml'
yaml_data = {'productos': productos}
yaml_path.write_text(yaml.dump(yaml_data, allow_unicode=True, sort_keys=False))
print("=== YAML ===")
print(yaml_path.read_text())
```

```python
# ── XML ─────────────────────────────────────────────────────
xml_path = OUTDIR / 'productos.xml'
root = ET.Element('productos')
for p in productos:
    prod_elem = ET.SubElement(root, 'producto')
    for k, v in p.items():
        child = ET.SubElement(prod_elem, k)
        child.text = str(v)

tree = ET.ElementTree(root)
ET.indent(tree, space='  ')  # Python 3.9+
tree.write(xml_path, encoding='unicode', xml_declaration=True)
print("=== XML ===")
print(xml_path.read_text())
```

```python
# ── JSON ────────────────────────────────────────────────────
json_path = OUTDIR / 'productos.json'
json_path.write_text(json.dumps({'productos': productos}, indent=2, ensure_ascii=False))
print("=== JSON ===")
print(json_path.read_text())
```

```python
# ── HTML ────────────────────────────────────────────────────
html_path = OUTDIR / 'productos.html'
html_path.write_text(df.to_html(index=False))
print("=== HTML ===")
print(html_path.read_text())
```

## Ítem 2 — Modificar el producto *jean* a tipo *skinny*

> **Modifique los archivos generados para especificar que el producto jean es de tipo skinny.**

```python
# Modificar el DataFrame
df.loc[df['desc_prod'] == 'jean', 'desc_prod'] = 'jean skinny'
productos[1]['desc_prod'] = 'jean skinny'

# Reescribir todos los archivos
df.to_csv(csv_path, sep='|', index=False)
txt_path.write_text(df.to_string(index=False))

yaml_data = {'productos': productos}
yaml_path.write_text(yaml.dump(yaml_data, allow_unicode=True, sort_keys=False))

root = ET.Element('productos')
for p in productos:
    prod_elem = ET.SubElement(root, 'producto')
    for k, v in p.items():
        child = ET.SubElement(prod_elem, k)
        child.text = str(v)
tree = ET.ElementTree(root)
ET.indent(tree, space='  ')
tree.write(xml_path, encoding='unicode', xml_declaration=True)

json_path.write_text(json.dumps({'productos': productos}, indent=2, ensure_ascii=False))
html_path.write_text(df.to_html(index=False))

print("Archivos actualizados con 'jean skinny'.")
print(df.to_string(index=False))
```

## Ítem 3 — Importar cada archivo y verificar

> **Importe cada uno de los archivos generados. Verifique que los datos se hayan cargado correctamente.**

```python
print("=== Lectura CSV ===")
df_csv = pd.read_csv(csv_path, sep='|', dtype={'id': str})
print(df_csv.to_string(index=False))
print(f"Shape: {df_csv.shape}, dtypes: {df_csv.dtypes.to_dict()}")
```

```python
print("=== Lectura TXT ===")
df_txt = pd.read_fwf(txt_path)
print(df_txt.to_string(index=False))
```

```python
print("=== Lectura YAML ===")
with open(yaml_path) as f:
    data_yaml = yaml.safe_load(f)
df_yaml = pd.DataFrame(data_yaml['productos'])
print(df_yaml.to_string(index=False))
```

```python
print("=== Lectura XML ===")
tree_in = ET.parse(xml_path)
root_in = tree_in.getroot()
registros = []
for prod in root_in.findall('producto'):
    registros.append({child.tag: child.text for child in prod})
df_xml = pd.DataFrame(registros)
print(df_xml.to_string(index=False))
```

```python
print("=== Lectura JSON ===")
with open(json_path) as f:
    data_json = json.load(f)
df_json = pd.DataFrame(data_json['productos'])
print(df_json.to_string(index=False))
```

```python
print("=== Lectura HTML ===")
df_html = pd.read_html(html_path)[0]
print(df_html.to_string(index=False))
```

## Conclusiones

| Formato | Librería de lectura | Observaciones |
|---------|--------------------|-----------------|
| CSV (`|`) | `pd.read_csv(sep='|')` | Más compacto y portable |
| TXT | `pd.read_fwf()` | Legible pero difícil de parsear automáticamente |
| YAML | `yaml.safe_load()` | Legible por humanos, soporta jerarquía |
| XML | `xml.etree.ElementTree` | Verbose, estándar en sistemas empresariales |
| JSON | `json.load()` | Estándar web, fácil de parsear |
| HTML | `pd.read_html()` | Útil para visualización en navegador |

Cada formato tiene sus casos de uso óptimos. CSV y JSON son los más comunes en ciencia de datos.
