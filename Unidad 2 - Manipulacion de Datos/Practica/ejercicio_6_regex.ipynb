## Ítem 1 — Validación de URL

> **Escriba una función que determine si una url es válida e imprima 'URL válida' o 'URL no válida'.**

```python
def validar_url(url):
    patron = re.compile(
        r'^(https?|ftp)://'          # protocolo
        r'([\w\-]+\.)+[\w\-]+'       # dominio
        r'(\.[a-zA-Z]{2,})'          # TLD
        r'(/[\w\-./?%&=#]*)?$'        # ruta opcional
    )
    if patron.match(url):
        print('URL válida')
    else:
        print('URL no válida')

# Pruebas
validar_url('https://pythondiario.com/')      # válida
validar_url('htps:/pythondiario.com/')        # no válida
validar_url('http://example.com/path?q=1')   # válida
validar_url('ftp://files.example.org')        # válida
validar_url('://missing-protocol.com')        # no válida
```

## Ítem 2 — Validación de email de Gmail

> **Escriba una función que determine si una dirección de correo electrónico es un correo electrónico de gmail válido.**

```python
def validar_gmail(email):
    # Gmail: letras, dígitos, puntos; no puede empezar/terminar con punto;
    # no puede tener dos puntos consecutivos
    patron = re.compile(
        r'^[a-zA-Z0-9]'
        r'([a-zA-Z0-9\.]*[a-zA-Z0-9])?'
        r'@gmail\.com$'
    )
    es_valido = bool(patron.match(email)) and '..' not in email
    estado = 'Gmail válido' if es_valido else 'Gmail no válido'
    print(f"{email!r:35s} → {estado}")

# Pruebas
validar_gmail('usuario@gmail.com')          # válido
validar_gmail('usuario.nombre@gmail.com')   # válido
validar_gmail('user123@gmail.com')           # válido
validar_gmail('.usuario@gmail.com')          # no válido (empieza con punto)
validar_gmail('usuario..nombre@gmail.com')   # no válido (doble punto)
validar_gmail('usuario@hotmail.com')         # no válido (no es gmail)
validar_gmail('usuario@gmail.com.ar')        # no válido
```

## Ítem 3 — Validación de fecha YYYY-MM-DD

> **Escriba una función que determine si un string corresponde a una fecha válida y se encuentra en el formato YYYY-MM-DD.**

```python
def validar_fecha(s):
    patron = re.compile(r'^\d{4}-(0[1-9]|1[0-2])-(0[1-9]|[12]\d|3[01])$')
    if not patron.match(s):
        print(f"{s!r} → Fecha no válida (formato incorrecto)")
        return
    # Verificar que la fecha sea real usando datetime
    from datetime import datetime
    try:
        datetime.strptime(s, '%Y-%m-%d')
        print(f"{s!r} → Fecha válida")
    except ValueError:
        print(f"{s!r} → Fecha no válida (fecha inexistente)")

# Pruebas
validar_fecha('2024-06-15')    # válida
validar_fecha('2024-02-29')    # válida (2024 es bisiesto)
validar_fecha('2023-02-29')    # no válida (2023 no es bisiesto)
validar_fecha('2024-13-01')    # no válida (mes 13)
validar_fecha('15-06-2024')    # no válida (formato incorrecto)
validar_fecha('2024-6-5')      # no válida (sin ceros)
```

## Ítem 4 — Análisis de la canción *Me gustas tú* de Manu Chao

> **(a) ¿Cuántas veces se hace referencia al verbo gustar?**
> **(b) ¿Cuántos verbos en infinitivo tiene la letra?**
> **(c) Lista de todas las cosas que le gustan a Manu Chao.**

```python
with open("../Datasets/Me_gustas_tu-Manu_Chao.txt", encoding='latin1') as f:
    letra = f.read()

print("--- Letra de la canción ---")
print(letra[:500], "...")
```

```python
# (a) Veces que aparece el verbo 'gustar' en cualquier conjugación
ocurrencias_gustar = re.findall(r'gust[a-záéíóú]*', letra, flags=re.IGNORECASE)
print(f"(a) Referencias al verbo 'gustar': {len(ocurrencias_gustar)}")
from collections import Counter
print("    Formas encontradas:", Counter(ocurrencias_gustar))
```

```python
# (b) Verbos en infinitivo: palabras terminadas en -ar, -er, -ir
infinitivos = re.findall(r'\b[a-záéíóúü]+(?:ar|er|ir)\b', letra, flags=re.IGNORECASE)
infinitivos_unicos = list(dict.fromkeys([v.lower() for v in infinitivos]))
print(f"(b) Verbos en infinitivo encontrados: {len(infinitivos)}")
print(f"    Únicos: {len(infinitivos_unicos)}")
print(f"    Lista: {infinitivos_unicos}")
```

```python
# (c) Cosas que le gustan a Manu Chao
# Patrón: 'Me gusta(n) X, me gustas tú'
patron_gusta = re.compile(
    r'[Mm]e gustan?\s+([^,\n]+?)\s*,\s*me gustas',
    flags=re.IGNORECASE
)
cosas = patron_gusta.findall(letra)

# Limpiar duplicados preservando orden
cosas_limpias = list(dict.fromkeys([c.strip().lower() for c in cosas]))

print(f"(c) Cosas que le gustan a Manu Chao ({len(cosas_limpias)} únicas):")
cosas_que_le_gustan = cosas_limpias
for i, cosa in enumerate(cosas_que_le_gustan, 1):
    print(f"  {i:2}. {cosa}")
```

## Conclusiones

- El verbo **gustar** y sus formas conjugadas aparecen varias decenas de veces en la canción.
- Se identificaron numerosos verbos en infinitivo (como *viajar*, *soñar*, *correr*, *volver*, *menear*, *camelar*).
- La lista de 'cosas que le gustan' incluye: los aviones, viajar, la mañana, el viento, soñar, la mar, la moto, correr, la lluvia, volver, marihuana, la montaña, la noche, la cena, la vecina, la guitarra, el reggae, la canela, el fuego, menear, La Coruña, Malasaña, la castaña, Guatemala, entre otros.
