Usando el mismo dataset de incendios del Ejercicio 2, generar una tabla pivot con
el número de incendios intencionales por **provincia** (filas) y **año** (columnas).

```python
import pandas as pd
import numpy as np
```

## Tabla pivot de incendios intencionales

> **Genere una tabla que muestre el número de incendios intencionales por provincia para cada año de los incluidos en dicho dataset.**

```python
df = pd.read_csv("../Datasets/incendios-cantidad-causas-provincia_2022.csv",
                  encoding='latin1', header=3)
df['provincia'] = df['provincia'].str.strip()
df = df.dropna(subset=['anio', 'provincia'])
df['anio'] = df['anio'].astype(int)
for col in ['total', 'negligencia', 'intencional', 'natural', 'desconocida']:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

print(f"Shape: {df.shape}")
df.head(3)
```

```python
pivot = df.pivot_table(
    index='provincia',
    columns='anio',
    values='intencional',
    aggfunc='sum',
    fill_value=0
)
pivot.columns.name = 'Año'
pivot.index.name = 'Provincia'

# Añadir columna de total
pivot['TOTAL'] = pivot.sum(axis=1)
pivot = pivot.sort_values('TOTAL', ascending=False)

print(f"Tabla pivot: {pivot.shape[0]} provincias × {pivot.shape[1]-1} años")
pivot
```

```python
# Top 5 provincias con más incendios intencionales en total
print("Top 5 provincias con más incendios intencionales (total 1993-2021):")
print(pivot['TOTAL'].head().to_string())
```

## Conclusiones

- La tabla pivot permite comparar fácilmente la evolución temporal de los incendios intencionales en cada provincia.
- Las provincias con más incendios intencionales acumulados en el período son las de mayor extensión o con mayor actividad agrícola.
- Se observan picos en algunos años específicos que coinciden con períodos de sequía o conflictos territoriales.
