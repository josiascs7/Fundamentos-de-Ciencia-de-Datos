```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Función de interpolación lineal por tramos

> **Construya una función que realice una interpolación lineal por tramos. La función debe recibir como entrada un valor x y devolver el valor interpolado de y, utilizando los intervalos definidos por los puntos dados.**

```python
# Puntos de referencia
PUNTOS = [(1, 2), (2, 3), (3, 5), (10, 6)]

def interpolar_lineal(x, puntos=PUNTOS):
    """Interpolación lineal por tramos.
    
    Para x fuera del rango, se devuelve el valor del extremo más cercano
    (extrapolación constante).
    """
    puntos = sorted(puntos, key=lambda p: p[0])
    xs = [p[0] for p in puntos]
    ys = [p[1] for p in puntos]

    if x <= xs[0]:
        return ys[0]
    if x >= xs[-1]:
        return ys[-1]

    for i in range(len(xs) - 1):
        if xs[i] <= x <= xs[i + 1]:
            t = (x - xs[i]) / (xs[i + 1] - xs[i])
            return ys[i] + t * (ys[i + 1] - ys[i])

# Pruebas
print("Pruebas de interpolación:")
casos = [1, 1.5, 2, 2.5, 3, 5, 6.5, 10]
for xv in casos:
    print(f"  x={xv:5.1f}  →  y={interpolar_lineal(xv):.4f}")
```

```python
# Visualización de la interpolación
xs_ref = [p[0] for p in PUNTOS]
ys_ref = [p[1] for p in PUNTOS]

xs_curva = np.linspace(0.5, 11, 500)
ys_curva = [interpolar_lineal(x) for x in xs_curva]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(xs_curva, ys_curva, color='steelblue', label='Interpolación lineal por tramos')
ax.scatter(xs_ref, ys_ref, color='firebrick', zorder=5, s=80, label='Puntos de referencia')
for x, y in PUNTOS:
    ax.annotate(f'({x},{y})', (x, y), textcoords='offset points', xytext=(5, 5))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Interpolación lineal por tramos')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Conclusiones

- La función `interpolar_lineal(x)` determina en qué segmento cae `x` y aplica la fórmula:
  $y = y_i + \dfrac{x - x_i}{x_{i+1} - x_i}(y_{i+1} - y_i)$
- Para valores fuera del rango $[1, 10]$ se retorna el valor del extremo más cercano.
- El tramo más pronunciado es entre $x=2$ y $x=3$ (pendiente = 2), mientras que entre $x=3$ y $x=10$ la pendiente es solo $\approx 0.14$.
