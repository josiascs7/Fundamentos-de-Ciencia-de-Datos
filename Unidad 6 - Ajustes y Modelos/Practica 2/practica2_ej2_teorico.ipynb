# Práctica 2 — Ejercicio 2: Regresión lineal múltiple (ejercicio teórico)

Supongamos que ajustamos un modelo de regresión lineal por mínimos cuadrados para el **precio de una propiedad** en función de los siguientes predictores:
- $x_1$: superficie en metros cuadrados
- $x_2$: cantidad de habitaciones
- $x_3$: antigüedad en años
- $x_4$: ubicación (0 = urbano, 1 = suburbano)

Con los siguientes coeficientes estimados:
$$\hat{\beta}_0 = 10000, \quad \hat{\beta}_1 = 5000, \quad \hat{\beta}_2 = 2000, \quad \hat{\beta}_3 = -1500, \quad \hat{\beta}_4 = -15000$$

---

```python
b0, b1, b2, b3, b4 = 10000, 5000, 2000, -1500, -15000
print("Modelo ajustado:")
print(f"  precio = {b0} + {b1}*x1 + {b2}*x2 + ({b3})*x3 + ({b4})*x4")

# Verificacion numerica para comparar urbano vs suburbano
sup, hab, ant = 80, 3, 10
p_urbano    = b0 + b1*sup + b2*hab + b3*ant + b4*0
p_suburbano = b0 + b1*sup + b2*hab + b3*ant + b4*1
print(f"\nEjemplo numerico (sup={sup}m2, hab={hab}, ant={ant} anos):")
print(f"  Precio urbano    (x4=0): ${p_urbano:,.0f}")
print(f"  Precio suburbano (x4=1): ${p_suburbano:,.0f}")
print(f"  Diferencia: ${p_urbano - p_suburbano:,.0f}")
```

## Ítem 1 — Ecuación del modelo en forma desarrollada

> **Escriba la ecuación del modelo ajustado en forma desarrollada.**

$$\hat{precio} = 10000 + 5000 \cdot x_1 + 2000 \cdot x_2 - 1500 \cdot x_3 - 15000 \cdot x_4$$

## Ítem 2 — Análisis de afirmaciones

> **A partir del modelo ajustado, ¿cuál de las siguientes afirmaciones es correcta y por qué?**
>
> **a)** Para un mismo número de habitaciones, antigüedad y metros cuadrados, las propiedades en áreas suburbanas son más caras, en promedio, que las que se encuentran en áreas urbanas.
>
> **b)** Para viviendas ubicadas en el área urbana con una misma antigüedad y superficie, el precio predicho por el modelo se incrementa, en promedio, en $10000 por cada año de antigüedad.
>
> **c)** El número de habitaciones es una variable que tiene "más peso" sobre el precio de la propiedad en viviendas en áreas urbanas que en aquellas que se encuentran en áreas suburbanas.
>
> **d)** El modelo predice que una propiedad ubicada en el área urbana cuesta $15000 más que una propiedad de la misma superficie, antigüedad y cantidad de habitaciones que se encuentre en el área suburbana.

---

**a) FALSA.** El coeficiente de x₄ es -15000. Una propiedad suburbana (x₄=1) tiene un precio
predicho $15000 MENOR que una urbana (x₄=0) con iguales características.

---

**b) FALSA.** El coeficiente de x₃ (antigüedad) es -1500, no +10000. Además, el efecto es negativo:
mayor antigüedad implica menor precio predicho (en $1500 por año adicional).

---

**c) FALSA.** No existe un término de interacción entre x₂ y x₄ en el modelo. El coeficiente β₂=2000
aplica por igual a propiedades urbanas y suburbanas; el modelo no distingue entre ambas para el
efecto de las habitaciones.

---

**d) CORRECTA.** El coeficiente β₄ = -15000 representa exactamente esta diferencia:

$$\hat{precio}_{urbano} - \hat{precio}_{suburbano} = -15000 \cdot (0-1) = +15000$$

La afirmación está correctamente formulada.
