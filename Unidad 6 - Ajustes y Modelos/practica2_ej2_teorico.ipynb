# Practica 2 - Ejercicio 2: Regresion lineal multiple (ejercicio teorico)

**Enunciado:** Modelo de regresion para precio de propiedad con coeficientes dados:
- β₀ = 10000, β₁ = 5000, β₂ = 2000, β₃ = -1500, β₄ = -15000

Donde: x₁ = superficie (m²), x₂ = habitaciones, x₃ = antiguedad (anos),
x₄ = ubicacion (0 = urbano, 1 = suburbano).

```python
b0, b1, b2, b3, b4 = 10000, 5000, 2000, -1500, -15000
print("Modelo ajustado:")
print(f"  precio = {b0} + {b1}*x1 + {b2}*x2 + ({b3})*x3 + ({b4})*x4")

# Verificacion numerica para comparar urbano vs suburbano
sup, hab, ant = 80, 3, 10
p_urbano    = b0 + b1*sup + b2*hab + b3*ant + b4*0
p_suburbano = b0 + b1*sup + b2*hab + b3*ant + b4*1
print(f"\nEjemplo numerico (sup={sup}m2, hab={hab}, ant={ant} anos):")
print(f"  Precio urbano    (x4=0): ${p_urbano:,.0f}")
print(f"  Precio suburbano (x4=1): ${p_suburbano:,.0f}")
print(f"  Diferencia: ${p_urbano - p_suburbano:,.0f}")
```

## 1. Ecuacion del modelo en forma desarrollada

$$\hat{precio} = 10000 + 5000 \cdot x_1 + 2000 \cdot x_2 - 1500 \cdot x_3 - 15000 \cdot x_4$$

## 2. Analisis de cada afirmacion

**a) Para un mismo numero de habitaciones, antiguedad y m², las propiedades suburbanas son mas caras.**

**FALSA.** El coeficiente de x₄ es -15000. Una propiedad suburbana (x₄=1) tiene un precio
predicho $15000 MENOR que una urbana (x₄=0) con iguales caracteristicas.

---

**b) Para viviendas urbanas con misma antiguedad y superficie, el precio se incrementa en $10000 por cada ano de antiguedad.**

**FALSA.** El coeficiente de x₃ (antiguedad) es -1500, no +10000. Ademas, el efecto es negativo:
mayor antiguedad implica menor precio predicho (en $1500 por ano adicional).

---

**c) El numero de habitaciones tiene mas peso sobre el precio en areas urbanas que suburbanas.**

**FALSA.** No existe un termino de interaccion entre x₂ y x₄ en el modelo. El coeficiente β₂=2000
aplica por igual a propiedades urbanas y suburbanas; el modelo no distingue entre ambas para el
efecto de las habitaciones.

---

**d) El modelo predice que una propiedad urbana cuesta $15000 mas que una suburbana de iguales caracteristicas.**

**CORRECTA.** El coeficiente β₄ = -15000 representa exactamente esta diferencia:

$$\hat{precio}_{urbano} - \hat{precio}_{suburbano} = -15000 \cdot (0-1) = +15000$$

La afirmacion esta correctamente formulada.
